```python
import cellrank as cr
import scanpy as sc
import scvelo as sv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sc.settings.set_figure_params(frameon=False, dpi=100)
cr.settings.verbosity = 2
```

```python
import warnings
warnings.simplefilter("ignore", category=UserWarning)
```

```python
summary_tab = pd.DataFrame(columns = ('Method', 'Dataset', 'Celltype', 'Mean Probability'))
```

```python
datasets = ['Pancreas_with_cc', 'DentateGyrus' , 'MouseBoneMarrow', 'MouseErythroid', 'HumanBoneMarrow']
methods = ['cell2fate', 'scvelo']
```

```python
correct_fates_dict = {'DentateGyrus': 
                 {'Astrocytes': ('Astrocytes', 'Radial_Glia-like'),
                 'OL': ('OPC', 'OL'),
                 'Granule_mature': ('Granule_mature', 'Radial Glia-like', 'Granule_immature', 'nIPC', 'Neuroblast')},
                 'Pancreas_with_cc':    
                 {'Alpha': ('Alpha'),
                 'Beta': ('Beta'),
                 'Delta': ('Delta'),
                 'Epsilon': ('Epsilon')},
                  'MouseErythroid':    
                 {'Erythroid3': ('Erythroid2', 'Erythroid1', 'Blood_progenitors_2', 'Blood_progenitors_1')},
                 'HumanBoneMarrow':    
                 {'CLP': ('CLP'),
                 'Mono_1': ('Mono_1', 'Mono_2'),
                 'DCs': ('DCs'),
                 'Ery_2': ('Ery_2', 'Ery_1'),
                 'Mega': ('Mega')},
                 'MouseBoneMarrow':                  
                 {'activating': ('activating', 'progenitors', 'dividing')}}
```

```python
for i in range(len(datasets)):
    for j in range(len(methods)):
        # Load data
        adata = sc.read_h5ad('/nfs/team283/aa16/data/fate_benchmarking/CellRank/' + methods[j] + datasets[i] + 'AnnDataForCellRank.h5ad')
        # Removing underscore makes cluster names compatabile with CellRank:
        adata.obs['clusters'] = [x.replace(" ", "+") for x in adata.obs['clusters']]
        # Run cellrank:
        from cellrank.kernels import VelocityKernel
        correct_fates = correct_fates_dict[datasets[i]]
        vk = VelocityKernel(adata, velocity_key="velocity")
        vk.compute_transition_matrix()
        vk.plot_projection()
        g = cr.estimators.GPCCA(vk)
        # Try finding as many macrostates as there are annotated clusters or else reduce number by 1 and try again
        n_states = int(np.round(len(np.unique(adata.obs['clusters']))*0.6))
        while n_states > 0:
            try:
                g.fit(cluster_key="clusters", n_states= n_states)
                n_states = 0
            except:
                n_states = n_states - 1
        macrostates = list(g.macrostates.value_counts().index)
        macrostates_split = [m.split('_')[0] for m in macrostates]
        print(macrostates)
        terminal_states = [macrostates[i] for i in range(len(macrostates)) if macrostates_split[i] in list(correct_fates.keys())]
        terminal_cells = [t.split('_')[0] for t in terminal_states]
        res = pd.DataFrame(columns = terminal_cells, index = ['Mean Probability'])
        res.loc['Mean Probability',:] = 0
        present_cells = []
        if len(terminal_states) > 1:
            g.set_terminal_states(states=terminal_states)
            g.compute_fate_probabilities(solver = 'direct')
            # Compute mean fate probability in cells belonging to a ground truth fate:
            for k in range(len(terminal_cells)):
                try:
                    adata.obs[terminal_cells[k].split('_')[0] + ' Probability']
                    adata.obs[terminal_cells[k] + ' Probability'] = np.array(g.fate_probabilities[terminal_cells[k]])
                    present_cells += [terminal_cells[k]]
                except:
                    adata.obs[terminal_cells[k] + ' Probability'] = 0
            for k in range(len(present_cells)):
                assigned_probability = adata.obs[present_cells[k] + ' Probability']
                true_fates = [a in np.array(correct_fates[present_cells[k]]) for a in adata.obs['clusters']]
                number_correct = np.sum(assigned_probability[true_fates])
                total_number = np.sum(true_fates)
                res.loc['Mean Probability', present_cells[k]] = np.round(number_correct/total_number,2)
        res['Mean'] = np.mean(res)
        if len(present_cells) > 0:
            fig, ax = plt.subplots(1, len(present_cells), figsize = (5*len(present_cells), 3))
            for k in range(len(present_cells)):
                if len(present_cells) > 1:
                    sc.pl.umap(adata, color = present_cells[k] + ' Probability', vmax = 0.6, ax = ax[k], show = False)
                else:
                    sc.pl.umap(adata, color = present_cells[k] + ' Probability', vmax = 0.6, ax = ax, show = False)
            plt.savefig('/nfs/team283/aa16/data/fate_benchmarking/benchmarking_results/CellRank/' + methods[j] + datasets[i] + '1.pdf')
        indices = [str(j) + str(i) + str(k) for k in range(len(res.columns))]
        summary_tab_temp = pd.DataFrame(columns = ('Method', 'Dataset', 'Celltype', 'Mean Probability'),
                                        index = indices)
        for k in range(len(res.columns)):
            summary_tab_temp.loc[indices[k],'Method'] = methods[j]
            summary_tab_temp.loc[indices[k],'Dataset'] = datasets[i]
            summary_tab_temp.loc[indices[k],'Celltype'] = res.columns[k]
            summary_tab_temp.loc[indices[k],'Mean Probability'] = res.loc['Mean Probability', res.columns[k]]
        summary_tab = pd.concat([summary_tab, summary_tab_temp])
        summary_tab.to_csv('/nfs/team283/aa16/data/fate_benchmarking/benchmarking_results/CellRank/CellRankResultsSummary1.csv')
```
